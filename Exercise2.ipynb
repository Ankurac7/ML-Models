Multivariate linear regression Codebasics

In the datasets folder (same level as this notebook on github) there is **hiring.csv**. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you we to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this we predict salaries for following candidates,

**2 yr experience, 9 test score, 6 interview score**

**12 yr experience, 10 test score, 10 interview score**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
df.experience=df.experience.fillna('zero')
df
```

```python
df.experience=df.experience.astype(str)
df.experience=df.experience.apply(w2n.word_to_num)
df
```

```python
import math
median_test_score=math.floor(df['test_score(out of 10)'].median())
test_score=df['test_score(out of 10)']
median_test_score
```

```python
df['test_score(out of 10)']=test_score.fillna(median_test_score)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']].values,df['salary($)'])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
